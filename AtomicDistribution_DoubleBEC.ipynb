```python
#Greiner map
greiner_data = {
    'red': ((0., 1, 1,), (.2, 0, 0), (.48, 0, 0), (.728, 1, 1),(0.912, 1, 1), (1, .5, .5)),
    'green': ((0., 1, 1), (.2, 0, 0), (.3, 0, 0), (.5, 1, 1),(.712, 1, 1), (.928, 0, 0), (1, 0, 0)),
    'blue': ((0., 1, 1), (.2, .5, .5), (.288, 1, 1), (.472, 1, 1),(.72, 0, 0), (1, 0, 0))
                }
```

```python
greiner = LinearSegmentedColormap('greiner',greiner_data)
#cmap=greiner
levels = MaxNLocator(nbins=15).tick_values(0.055,0.55)
norm = BoundaryNorm(levels, ncolors=greiner.N, clip=True)
```

```python
for lst_n in lsts:
    N = len(lst_n)
    A = []
    for file in lst_n:
        pic= plt.imread('K39/'+file)
        A.append(pic)
    Avr = sum(A)/N

    
    tm = str(int(float(re.split('StopEvaporation_|.png',lst_n[0])[1])*100))
    print(tm)
    np.savetxt('K39/Average_Matrices/Average_39_{}.txt'.format(tm),Avr,fmt='%.18e',delimiter=' ')
    
    fig, ax = plt.subplots(figsize=(6,6))
    #ax.imshow(Avr[100:441,120:461],cmap=plt.cm.gray,norm=norm)
    ax.imshow(Avr,cmap=plt.cm.gray,norm=norm)
    plt.savefig('K39/Average_Matrices/Img_39_{}.png'.format(tm),dpi=300)
```
